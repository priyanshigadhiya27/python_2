```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame({"a":[1,2],"b":[3,4]})
df2=pd.DataFrame({"c":[5,7],"d":[7,9]})
print(df1)
print(df2)
pd.concat((df1,df2),ignore_index=True)

```

```python
pd.concat((df1,df2),axis=1)
```

```python
df1=pd.DataFrame({"a":[1,2],"b":[3,4]})
df2=pd.DataFrame({"a":[5,7,8],"b":[7,9,78]})
print(df1)
print(df2)
pd.concat((df1,df2),axis=0)

```

```python
pd.concat((df1,df2))
```

```python
df1=pd.DataFrame({"a":[1,2],"b":[3,4]})
df2=pd.DataFrame({"a":[5,7],"b":[7,9]})
print(df1)
print(df2)
pd.concat((df1,df2),ignore_index=False,keys=["A1","B1"])
```

```python
df1=pd.DataFrame({"b":[1,2],"a":[3,4]})
df2=pd.DataFrame({"c":[5,7],"b":[7,9]})
print(df1)
print(df2)
pd.concat((df1,df2),join="inner")
```

```python
pd.concat((df1,df2),join="outer",sort=True)
```

```python
df1=pd.DataFrame({'sub':['FSD','Python','DM'],'fac':['nas','vha','mss'],"id_x":[1,2,4]})
df2=pd.DataFrame({'id_y':[1,2,3],'mark':[9,10,11]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="inner") #commom id  
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
df1.merge(df2,on="id")
```

```python
df1.merge(df2,on="id",how="left")
```

```python
df1.merge(df2,on="id",how="right")
```

```python
df1.merge(df2,how="cross")
```

```python
df1.merge(df2,left_on="id_x",right_on="id_v",left_index=True,right_index=True)
```

```python
df1=pd.DataFrame({'sub':['FSD','Python','DM'],'fac_x':['nas','vha','mss'],"id_x":[1,2,4]})
df2=pd.DataFrame({'id_y':[1,2,3],'mark':[9,10,11],'fac_y':['nas','vha','mss']})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2,left_on="id_x",right_on="id_y",suffixes=("_df1","_df2"))
```
