```python
import numpy as np
import pandas as pd
```

```python
# import like this
items_path = "items.csv"
q1_path = "quarter-1.csv"
q2_path = "quarter-2.csv"
q3_path = "quarter-3.csv"


q1= pd.read_csv(q1_path)
q2 = pd.read_csv(q2_path)
q3 = pd.read_csv(q3_path)

items = pd.read_csv(items_path)
```

```python
df=pd.concat((q1,q2,q3),keys=["Q1","Q2","Q3"])
df
```

```python
df=df.reset_index()
df
```

```python
df=df.merge(items,on="item_id")
df
```

```python
df.groupby(['level_0','item_name'],as_index=False)['quantity'].sum().sort_values('quantity',ascending=False).drop_duplicates(subset=['level_0'])
```

```python
df["price"]=df['item_price'].apply(lambda x: float(x[1:]))
df["price"]
```

```python
df["total_price"]=df["price"]*df["quantity"]

```

```python
df
```

```python
df.groupby(["level_0","item_name"],as_index=False)["total_price"].sum().sort_values("total_price",ascending=False).drop_duplicates(subset=["level_0"])
```
