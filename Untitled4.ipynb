```python
import plotly.express as px
import numpy as np
import pandas as pd
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
df=pd.read_csv("social_media_engagement.csv.")
df
```

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,height=500,width=800,title='hello world',color='platform',
          color_discrete_map={'Facebook':'red','Instgram':'blue','Twitter':'green'})
fig.show()
```

```python
px.bar(data_frame=df,x='post_type',y=['likes','comments','shares'],color='platform',text='comments',barmode='group',
      pattern_shape='sentiment_score')
```

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,height=1500,width=2000,title='hello world',color='platform',
          color_discrete_map={'Facebook':'red','Instgram':'blue','Twitter':'green'},facet_col='sentiment_score',facet_row='post_day')
fig.show()
```

```python
px.scatter(data_frame=df,x='likes',y='comments',color='platform',symbol='platform',size='shares',
          facet_col='sentiment_score',facet_row='post_day')
```

```python
px.scatter_3d(data_frame=df,x='likes',y='comments',z='shares',height=1000,width=700,color='platform')
```

```python
px.scatter_matrix(data_frame=df,dimensions=['likes','comments','shares'],color='platform',size='likes',)
```

```python
px.line(data_frame=df,x='post_id',y='likes',color='post_type',hover_data='sentiment_score',markers=True)
```
